```python
xmas_counts = [s.count("XMAS") for s in input_string]
samx_counts = [s.count("SAMX") for s in input_string]

print("XMAS counts:", sum(xmas_counts))
print("SAMX counts:", sum(samx_counts))
```

```python
new_list = []
for line in input_string:
    line = list(line)
    new_list.append(line)

```

```python
counter = 0

for i in range(len(new_list) - 3):
    for j in range(len(new_list[i])):
        if new_list[i][j] == 'X':
            if j < len(new_list[i + 1]) and new_list[i + 1][j] == 'M':
                if j < len(new_list[i + 2]) and new_list[i + 2][j] == 'A':
                    if j < len(new_list[i + 3]) and new_list[i + 3][j] == 'S':
                        counter += 1

print("Counter:", counter)
```

```python
counter_samx = 0

for i in range(len(new_list) - 3):
    for j in range(len(new_list[i])):
        if new_list[i][j] == 'S':
            if j < len(new_list[i + 1]) and new_list[i + 1][j] == 'A':
                if j < len(new_list[i + 2]) and new_list[i + 2][j] == 'M':
                    if j < len(new_list[i + 3]) and new_list[i + 3][j] == 'X':
                        counter_samx += 1

print("Counter SAMX:", counter_samx)
```

```python
counter_diagonal_xmas = 0

for i in range(len(new_list) - 3):
    for j in range(len(new_list[i]) - 3):
        if new_list[i][j] == 'X':
            if new_list[i + 1][j + 1] == 'M':
                if new_list[i + 2][j + 2] == 'A':
                    if new_list[i + 3][j + 3] == 'S':
                        counter_diagonal_xmas += 1

print("Counter Diagonal XMAS:", counter_diagonal_xmas)
```

```python
counter_diagonal_samx = 0

for i in range(len(new_list) - 3):
    for j in range(len(new_list[i]) - 3):
        if new_list[i][j] == 'S':
            if new_list[i + 1][j + 1] == 'A':
                if new_list[i + 2][j + 2] == 'M':
                    if new_list[i + 3][j + 3] == 'X':
                        counter_diagonal_samx += 1

print("Counter Diagonal SAMX:", counter_diagonal_samx)
```

```python
counter_other_diagonal_xmas = 0
counter_other_diagonal_samx = 0

for i in range(3, len(new_list)):
    for j in range(len(new_list[i]) - 3):
        if new_list[i][j] == 'X':
            if new_list[i - 1][j + 1] == 'M':
                if new_list[i - 2][j + 2] == 'A':
                    if new_list[i - 3][j + 3] == 'S':
                        counter_other_diagonal_xmas += 1
        if new_list[i][j] == 'S':
            if new_list[i - 1][j + 1] == 'A':
                if new_list[i - 2][j + 2] == 'M':
                    if new_list[i - 3][j + 3] == 'X':
                        counter_other_diagonal_samx += 1

print("Counter Other Diagonal XMAS:", counter_other_diagonal_xmas)
print("Counter Other Diagonal SAMX:", counter_other_diagonal_samx)
```

```python
counter_total = counter + counter_samx + counter_diagonal_xmas + counter_diagonal_samx + counter_other_diagonal_xmas + counter_other_diagonal_samx + sum(xmas_counts) + sum(samx_counts)
print("Total Counter:", counter_total)
```
